```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    ## Remove all the special characters otherthen small a-z and capital A-Z
    review = review.lower()
    ## lowering all the sentences Duplicates words are not be present in there
    review = review.split()
    ## Spliting of the entire words
    ## we don't use this stopword we use lemmatizer ---> # review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    ## So we use Lemetization
    review = [lemmatizer.lemmatize(word) for word in review]
    ## each and every words we traverse and apply stopwords and then we applying stemming

    review = ' '.join(review)
    ## join the review
    corpus.append(review)
    ## append on corpus list
```
