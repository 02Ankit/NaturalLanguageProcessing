```python
## in the paragraph there are so many special charecters 
## so we should clean this special charector
import re    
```

```python
corpus=[] ## new corpus / paragraph
for i in range(len(sentences)):
    ## replace the special character
    ## replace karenge subhi special characters ko small a-z and capital A-Z ko chod ke 
    ## write a condition a-z and A-Z ko chod kar subhi special character space ya blank me convert ho jayega 
    review = re.sub('[^a-zA-Z]', ' ', sentences[i] )
    ## convert karenge subhi character ko lower me  
    review = review.lower()
    corpus.append(review)
```

```python
## Stemming
for i in corpus: ## iterate every corpus or sentence
    words = nltk.word_tokenize(i) ## subhi sentences ke unique words ko lenge 
    for word in words: ## every word iterate 
        ## stopword english ke sare common words ko corpus se leta hai
        if word not in set(stopwords.words('english')): ## jo bhi words stopword me present nahi hai unhe le rahe hai.     
            print(stemmer.stem(word))
    
```

```python
## Apply Stopwords. lemmatize
## basically if we remove the words like a the we as in use like more
## we should use stopwords
import re 
corpus =[]
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
## Using Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
## there are lot of features in countvectorizer
cv = CountVectorizer(binary = True, ngram_range=(3,3))
```

##### binary = True is liye karenge kyunki jub bag of words ke index check karne me 
##### ngrams_range =(3,3)----> there will be only triGrams  
##### ngrams_range =(2,3)----> it will start with BiGrams after getting Bigram it will take TryGram  
##### repeated words ko count 0,1,or 2,3... bhi karta hai is liye Binary Bag of Word use kiye hai 
##### binary = True ka matlab Binary Bag of Word hota hai 

```python
X = cv.fit_transform(corpus) ## corpus ko CountVectorize me fit karenge 
```
